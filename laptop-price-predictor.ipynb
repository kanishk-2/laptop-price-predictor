```python
# Assuming your DataFrame is named 'df' and has a column named 'Memory'

# Make sure 'df' is defined and contains the 'Memory' column before running this code.

# --- Feature Engineering for 'Memory' Column ---
df['Memory'] = df['Memory'].astype(str).replace(r'\.0', '', regex=True)
df["Memory"] = df["Memory"].str.replace('GB', '')
df["Memory"] = df["Memory"].str.replace('TB', '000')
new = df["Memory"].str.split("+", n = 1, expand = True)

df["first"]= new[0].str.strip()
df["second"]= new[1].fillna("0")

# Function to extract storage type and size
def extract_storage_info(series):
    hdd = 0
    ssd = 0
    hybrid = 0
    flash_storage = 0
    for item in series:
        item = str(item).strip()
        size_str = ''.join(filter(str.isdigit, item))
        if size_str:
            size = int(size_str)
            if "HDD" in item:
                hdd += size
            elif "SSD" in item:
                ssd += size
            elif "Hybrid" in item:
                hybrid += size
            elif "Flash Storage" in item:
                flash_storage += size
    return pd.Series([hdd, ssd, hybrid, flash_storage])

storage_info_first = df["first"].apply(lambda x: [x]).apply(extract_storage_info)
df["HDD_first"] = storage_info_first[0]
df["SSD_first"] = storage_info_first[1]
df["Hybrid_first"] = storage_info_first[2]
df["Flash_Storage_first"] = storage_info_first[3]

storage_info_second = df["second"].apply(lambda x: [x]).apply(extract_storage_info)
df["HDD_second"] = storage_info_second[0]
df["SSD_second"] = storage_info_second[1]
df["Hybrid_second"] = storage_info_second[2]
df["Flash_Storage_second"] = storage_info_second[3]

df["HDD"] = df["HDD_first"] + df["HDD_second"]
df["SSD"] = df["SSD_first"] + df["SSD_second"]
df["Hybrid"] = df["Hybrid_first"] + df["Hybrid_second"]
df["Flash_Storage"] = df["Flash_Storage_first"] + df["Flash_Storage_second"]

df.drop(columns=['Memory', 'first', 'second', 'HDD_first', 'SSD_first', 'Hybrid_first',
                'Flash_Storage_first', 'HDD_second', 'SSD_second', 'Hybrid_second',
                'Flash_Storage_second'], inplace=True)

# Now your DataFrame 'df' should have the 'HDD', 'SSD', 'Hybrid', and 'Flash_Storage' columns
# representing the total storage of each type in GB.
```

### Voting Regressor

```python
from sklearn.ensemble import VotingRegressor

# Define the column transformer
step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
], remainder='passthrough')

# Define individual regressors
rf = RandomForestRegressor(n_estimators=350, random_state=3, max_samples=0.5, max_features=0.75, max_depth=15, bootstrap=True)
gbdt = GradientBoostingRegressor(n_estimators=100, max_features=0.5)
xgb = XGBRegressor(n_estimators=25, learning_rate=0.3, max_depth=5)
et = ExtraTreesRegressor(n_estimators=100, random_state=3, max_samples=0.5, max_features=0.75, max_depth=10, bootstrap=True)

# Define the Voting Regressor
step2 = VotingRegressor([('rf', rf), ('gbdt', gbdt), ('xgb', xgb), ('et', et)], weights=[5, 1, 1, 1])

# Create the pipeline
pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

# Fit the pipeline and make predictions
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)

# Print evaluation metrics
print('R2 score', r2_score(y_test, y_pred))
print('MAE', mean_absolute_error(y_test, y_pred))
```

### Stacking

```python
from sklearn.ensemble import StackingRegressor
#  Define the column transformer
step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
], remainder='passthrough')

# Define base estimators
estimators = [
    ('rf', RandomForestRegressor(n_estimators=350, random_state=3, max_samples=0.5, max_features=0.75, max_depth=15)),
    ('gbdt', GradientBoostingRegressor(n_estimators=100, max_features=0.5)),
    ('xgb', XGBRegressor(n_estimators=25, learning_rate=0.3, max_depth=5))
]

# Define the final estimator
step2 = StackingRegressor(estimators=estimators, final_estimator=Ridge(alpha=100))

# Create the pipeline
pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

# Fit the pipeline and make predictions
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)

# Print evaluation metrics
print('R2 score', r2_score(y_test, y_pred))
print('MAE', mean_absolute_error(y_test, y_pred))
```
